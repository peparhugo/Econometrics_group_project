```python
import pandas as pd
import numpy as np 
import plotly.express as px
```

```python
df= pd.read_csv('JPM.csv')
```

```python
# df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
fig = px.line(df, x=df['Date'], y=df['Adj Close'])
fig.show()
```

```python
df['Adj Close'].describe(percentiles=[0.05,0.1,0.25,.33,0.5,0.67,0.75,0.9,0.95,0.99])
```

```python
df['avg_stock_value']=(df['Adj Close'].iloc[0]+df['Adj Close'].iloc[-1])/2
df['mean_stock_value']=df['Adj Close'].mean(0)
```

```python
df['Volatility']= ((df['Adj Close']-df['mean_stock_value'])**2)/df.shape[0]
```

```python
df
```

```python
fig = px.line(df, x=df['Date'], y=df['Volatility'])
fig.show()
```

```python
df['Daily Return'] = df['Adj Close']/(df['Adj Close'].shift(1)-1)
fig = px.line(df, x=df['Date'], y=df['Daily Return'])
fig.show()
```
